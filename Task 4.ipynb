```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
file_path = "C:/Users/Harishma/Downloads/Internship/Task 4/matches.csv"
ipl_df = pd.read_csv(file_path)
ipl_df
```

```python
print(ipl_df.head())
print(ipl_df.info())
```

```python
# Check for missing values
print(ipl_df.isnull().sum())

# Fill or drop missing values as appropriate
ipl_df = ipl_df.fillna({'umpire3': 'Unknown'})
ipl_df = ipl_df.dropna()
```

```python
print(ipl_df.describe())
```

# Team Performance Analysis:

```python
# Count of wins per team
wins_per_team = ipl_df['winner'].value_counts()

# Plotting wins per team
import matplotlib.pyplot as plt
wins_per_team.plot(kind='bar', title='Number of Wins per Team')
plt.xlabel('Teams')
plt.ylabel('Number of Wins')
plt.show()
```

# Player Performance Analysis:

```python
# Top players by Player of the Match awards
top_players = ipl_df['player_of_match'].value_counts().head(10)

# Plotting top players
top_players.plot(kind='bar', title='Top Players by Player of the Match Awards')
plt.xlabel('Players')
plt.ylabel('Number of Awards')
plt.show()
```

# Factors Contributing to Wins/Losses:

```python
# Analyzing the impact of toss on match results
toss_wins = ipl_df[ipl_df['toss_winner'] == ipl_df['winner']]

# Percentage of matches won by the team that won the toss
toss_win_percentage = len(toss_wins) / len(ipl_df) * 100

# Display toss win percentage
print(f"Percentage of matches won by the team that won the toss: {toss_win_percentage:.2f}%")
```

# Win by Runs vs. Win by Wickets:

```python
# Plotting win by runs vs win by wickets
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(ipl_df['win_by_runs'][ipl_df['win_by_runs'] > 0], bins=30, color='blue', alpha=0.7)
plt.title('Distribution of Wins by Runs')
plt.xlabel('Runs')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.hist(ipl_df['win_by_wickets'][ipl_df['win_by_wickets'] > 0], bins=30, color='green', alpha=0.7)
plt.title('Distribution of Wins by Wickets')
plt.xlabel('Wickets')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

# Home Ground Advantage:

```python
# Analyzing home ground advantage
ipl_df['is_home_team_winner'] = ipl_df.apply(lambda row: 1 if (row['city'] in row['team1'] and row['winner'] == row['team1']) or (row['city'] in row['team2'] and row['winner'] == row['team2']) else 0, axis=1)
home_win_percentage = ipl_df['is_home_team_winner'].mean() * 100

# Display home win percentage
print(f"Percentage of matches won by the home team: {home_win_percentage:.2f}%")
```

# Win Distribution by Season

```python
import seaborn as sns

# Win distribution by season for each team
plt.figure(figsize=(14, 7))
sns.countplot(data=ipl_df, x='season', hue='winner', palette='Set2')
plt.title('Win Distribution by Season')
plt.xlabel('Season')
plt.ylabel('Number of Wins')
plt.legend(title='Teams', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

# Top Venues by Number of Matches

```python
# Top venues by number of matches
top_venues = ipl_df['venue'].value_counts().head(10)

# Plotting top venues
plt.figure(figsize=(14, 7))
sns.barplot(x=top_venues.values, y=top_venues.index, palette='viridis')
plt.title('Top Venues by Number of Matches')
plt.xlabel('Number of Matches')
plt.ylabel('Venue')
plt.show()
```

# Impact of Toss Decision

```python
# Toss decision impact on match result
toss_decision_wins = ipl_df[ipl_df['toss_winner'] == ipl_df['winner']]['toss_decision'].value_counts()

# Plotting toss decision impact
plt.figure(figsize=(7, 7))
plt.pie(toss_decision_wins, labels=toss_decision_wins.index, autopct='%1.1f%%', startangle=140, colors=['lightcoral', 'lightskyblue'])
plt.title('Impact of Toss Decision on Match Result')
plt.show()
```

# Most Common Match Outcomes

```python
# Match result distribution
match_result_dist = ipl_df['result'].value_counts()

# Plotting match result distribution
plt.figure(figsize=(14, 7))
sns.barplot(x=match_result_dist.index, y=match_result_dist.values, palette='plasma')
plt.title('Most Common Match Outcomes')
plt.xlabel('Match Result')
plt.ylabel('Number of Matches')
plt.show()
```

# Top Umpires

```python
# Top umpires by number of matches
top_umpires = pd.concat([ipl_df['umpire1'], ipl_df['umpire2']]).value_counts().head(10)

# Plotting top umpires
plt.figure(figsize=(14, 7))
sns.barplot(x=top_umpires.values, y=top_umpires.index, palette='magma')
plt.title('Top Umpires by Number of Matches')
plt.xlabel('Number of Matches')
plt.ylabel('Umpire')
plt.show()
```

The Indian Premier League (IPL) is one of the most popular and competitive cricket tournaments globally. This analysis aims to uncover the most successful teams, highlight top-performing players, and identify key factors influencing match outcomes. 

Success Metrics: Teams and Players

Most Successful Teams

By analyzing the number of wins each team has accumulated over the seasons, the Mumbai Indians (MI) and Chennai Super Kings (CSK) stand out as the most victorious teams in IPL history. MI's consistent performance and strategic gameplay have led them to multiple championships. Similarly, CSK, led by the astute captain MS Dhoni, has demonstrated remarkable consistency and resilience, securing numerous wins and titles.

Top Players

Individual brilliance often sways match outcomes. Analysis of 'Player of the Match' awards reveals that AB de Villiers, Chris Gayle, and MS Dhoni frequently emerge as game-changers. AB de Villiers' exceptional batting skills and versatility have earned him numerous accolades. Chris Gayle is renowned for his explosive batting, while MS Dhoni's leadership and finishing abilities have won him critical games, cementing his reputation as a reliable player in high-pressure situations.

Key Factors Influencing Match Outcomes

Impact of Toss

The toss plays a crucial role in determining match outcomes. Teams winning the toss and making informed decisions about batting or fielding first often have a higher winning percentage. Specifically, teams choosing to bat first can leverage favorable pitch conditions and set imposing targets.

Win by Runs vs. Win by Wickets

Examining the distribution of wins by runs and wickets provides insights into team strategies. Teams like MI and CSK often secure victories by significant margins, whether by piling up runs or through strategic bowling to restrict opponents and achieve wins by wickets.

Home Ground Advantage

Home ground advantage is a well-known phenomenon in sports, and the IPL is no exception. Teams playing at their home venues often benefit from familiar pitch conditions and local fan support, increasing their probability of winning.
